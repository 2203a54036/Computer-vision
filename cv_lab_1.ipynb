```python
!pip install opencv-python

import cv2
import matplotlib.pyplot as plt

from google.colab import files
uploaded = files.upload()
file_read=list(uploaded.keys())[0]
img=cv2.imread(file_read)
```

```python
colorchanger= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(colorchanger)
plt.axis('off')
plt.title("BGR TO RGB")
```

```python
colorchanger= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(colorchanger, cmap='gray')
plt.axis('off')
plt.title("BGR TO GRAYSCALE")
```

```python
vflip = cv2.flip(colorchanger, 0)
plt.imshow(vflip, cmap='gray')
plt.axis('off')
plt.title("Flipped Vertically")
plt.show()
```

```python
r90 = cv2.rotate(colorchanger, cv2.ROTATE_90_CLOCKWISE)
plt.imshow(r90, cmap='gray')
plt.axis('off')
plt.title("Rotated 90")
plt.show()
```

```python

resized_img = cv2.resize(img, (300, 100)) # Resize to 300x200 pixels
plt.imshow(cv2.cvtColor(resized_img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Resized Image")
plt.show()
```
